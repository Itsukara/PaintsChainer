```python
import cv2, matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pic_name = "IMG_6731"
bur_size = 7
threshold1=20
threshold2=60

pic = cv2.imread(pic_name + ".jpg")
pic_preprocessed  = cv2.cvtColor(cv2.GaussianBlur(pic, (bur_size,bur_size), 0), cv2.COLOR_BGR2GRAY)
pic_edges = cv2.bitwise_not(cv2.Canny(pic_preprocessed, threshold1=threshold1, threshold2=threshold2))
plt.imshow(cv2.cvtColor(pic_edges, cv2.COLOR_GRAY2RGB))
cv2.imwrite(pic_name + "_edges.jpg", pic_edges)
```
